```python
class Punto():

  def __init__(self, x, y):
    self.x = x
    self.y = y

  def eje_x(self):
    return self.x

  def eje_y(self):
    return self.y

  def impresion(self):
    return f"El eje x es {str(self.x)} + y el eje y es {str(self.y)}"

  def opuesto(self):
    return Punto(-self.x, -self.y)

class Linea():

  def __init__(self, punto_a, punto_b):
    self.punto_a = punto_a
    self.punto_b = punto_b

  def mueve_derecha(self):
    self.punto_a.x += 1
    self.punto_b.x += 1

  def mueve_izquierda(self):
    self.punto_a.x -= 1
    self.punto_b.x -= 1

  def mueve_arriba(self):
    self.punto_a.y += 1
    self.punto_b.y += 1

  def mueve_abajo(self):
    self.punto_a.y -= 1
    self.punto_b.y -= 1

p1 = Punto(1, 2)
p2 = Punto(3, 4)

linea = Linea(p1, p2)

print(p1.impresion())
print(p2.impresion())

linea.mueve_derecha()
print("Si se mueve el eje x 1 vez a la derecha los nuevos ejes son:")
print(p1.impresion())
print(p2.impresion())

linea.mueve_arriba()
print("Si se mueve el eje y 1 vez hacia arriba los nuevos ejes son:")
print(p1.impresion())
print(p2.impresion())


```
